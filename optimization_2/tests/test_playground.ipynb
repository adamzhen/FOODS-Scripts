```python
import numpy as np
import matplotlib.pyplot as plt
```

### Effect of W4 on SAV

```python
# Read data from W4_SAV_Effect.txt
data = np.loadtxt('W4_SAV_Effect.txt', skiprows=1, delimiter=',')

# Plot the data
plt.plot(data[:,0], data[:,-1], 'o-')
plt.xlabel('W4/W3 Ratio')
plt.ylabel('SAV (1/cm)')
plt.title('Effect of W4/W3 Ratio on SAV')
plt.show()
```
